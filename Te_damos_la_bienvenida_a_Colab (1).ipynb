```
#Calculadora IMC
peso = float(input("ingresa tu peso"))
altura = float(input("ingresa tu altura"))

imc = peso / (altura ** 2)
print(f"Tu IMC es: {imc:.2f}")

if imc < 18.5:
  print("Bajo peso")
elif imc < 25:
  print("Peso normal")
elif imc < 30:
  print("Sobre peso")
else:
  print("Eres un kilo mortal")
```
